# Relatório de Análises VI

## Criando Agrupamentos

## Média do "valor bruto" apenas entre os bairros

```python
dados['Valor'].mean()
```

```python
bairro1 = ['Barra da Tijuca', 'Botafogo', 'Copacabana', 'Flamengo', 'Ipanema', 'Leblon', 'Tijuca']
selecao1 = dados['Bairro'].isin(bairro1)
dados1 = dados[selecao1]
dados1
```

```python
dados1['Valor Bruto'].mean()
```

```python
dados1['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados1.groupby('Bairro')
```

```python
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, dados in grupo_bairro:
  print('{} -> {}'.format(bairro, dados['Valor'].mean()))
```

## Estatíticas Descritivas

```python
grupo_bairro['Valor'].describe()
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo'})
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('figure', figsize = (15, 7))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel', {'fontsize': 12})
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```

# Gráficos

```python
dados = pd.read_csv('/aluguel.csv', sep = ';')
```

```python
dados.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados['Valor'], dados['Area'])
g1.set_title('Valor x Área')
```

```python
area
```

```python
g2.hist(dados['Valor'])
g2.set_title('Histograma')
```

```python
dados_g3 = dados['Valor'].sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra Valor')
```

```python
grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Médio por Tipo')
```
